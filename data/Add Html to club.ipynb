```python
import urllib
import urllib.request
import urllib.parse
from urllib.request import urlopen 
from urllib.request import urlretrieve 


from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
```

```python
df = pd.read_csv('Club_geo.csv')
```

```python
searckLink = 'http://badiste.fr/liste-club-badminton?sigle=&nom=Zem%27bad+Le+Montat&ville=&Submit=Rechercher'
html = []
errors = []
for i, row in df.iterrows():
    if(i%100 == 0):
        print(i)
    name = row['name']
    uc = name.encode('ISO-8859-1')
    data = urllib.parse.urlencode({'nom': uc})
    url = 'http://badiste.fr/liste-club-badminton?sigle=&'+ data +'&ville=&Submit=Rechercher'
    try:
        req = urllib.request.Request(url)
        with urllib.request.urlopen(req) as f:
            resp = f.read()

        soup = BeautifulSoup(resp, 'html.parser')
        s = soup.find("div", { "id" : "tabcontent" })
        html.append(s.find_all("div")[1])
    except AttributeError:
        html.append('')
        errors.append(name)
            

    
    
```

```python
df['html'] = html
df.to_csv('Club_geo.csv')
```
